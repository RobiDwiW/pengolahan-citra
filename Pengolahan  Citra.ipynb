```python
# import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Membaca Gambar
img = cv2.imread("anis.Jpeg")
```

```python
# Menampilkan Gambar 
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
cvImg = cv2.multiply(img,2)
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
npImg = img*4
plt.imshow(npImg[:,:,::-1])
plt.show()
```
